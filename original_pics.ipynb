```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os 
import time

import PIL
from PIL import Image
```

```python
path =  '/Users/CagriCivici/Desktop/python_ws/data_augmentation/original_pic/'
imgs = []
```

```python
valid_images = [".jpg",".gif",".jpeg",".tga"]
for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_images:
        continue
    imgs.append(Image.open(os.path.join(path,f)))
```

```python
pics=len(imgs)
col = 5
row = pics / col
axes =[]
fig=plt.figure()
resizer = []
for element in(imgs):
    img = element.resize((60,80), Image.ANTIALIAS)
    resizer.append(img)
    print(element.filename)
    #plt.imshow(img)
    

for i in range(pics):
    pic_print = resizer[i]
    axes.append( fig.add_subplot(row, col, i+1) )
    #subplot_title=("Subplot"+str(i))
    #axes[-1].set_title(subplot_title) 
    plt.imshow(pic_print)
    
fig.tight_layout(pad=0.2, w_pad=0.1, h_pad=0.1) 
plt.savefig('/Users/CagriCivici/Desktop/python_ws/data_augmentation/readME_picture.jpeg')
plt.show()


    
   
```
